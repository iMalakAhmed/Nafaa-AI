```python
import subprocess, os, glob

audio_full_dir = os.path.join(base_dir, "audio_full")
os.makedirs(audio_full_dir, exist_ok=True)

mp4s = sorted(glob.glob(os.path.join(video_dir, "*.mp4")))
if not mp4s:
    raise FileNotFoundError("No MP4 found. Check download output.")
mp4_path = mp4s[0]

full_wav = os.path.join(audio_full_dir, "full.wav")
subprocess.run([
    "ffmpeg", "-y",
    "-i", mp4_path,
    "-vn",
    "-ac", "1",
    "-ar", "16000",
    full_wav
], check=True)

print("MP4:", mp4_path)
print("Full WAV:", full_wav)
```

```python
import math, os, subprocess

clips_audio_dir = os.path.join(base_dir, "audio_15s")
os.makedirs(clips_audio_dir, exist_ok=True)

dur = float(subprocess.check_output([
    "ffprobe", "-v", "error",
    "-show_entries", "format=duration",
    "-of", "default=noprint_wrappers=1:nokey=1",
    full_wav
]).decode().strip())

segment = 15
n = math.ceil(dur / segment)

def fmt_hhmmss(t):
    t = int(t)
    h = t // 3600
    m = (t % 3600) // 60
    s = t % 60
    return f"{h:02d}{m:02d}{s:02d}"  # HHMMSS

for i in range(n):
    start = i * segment
    end = min((i + 1) * segment, int(math.ceil(dur)))
    name = f"{fmt_hhmmss(start)}_{fmt_hhmmss(end)}"
    out_wav = os.path.join(clips_audio_dir, f"{name}.wav")

    subprocess.run([
        "ffmpeg", "-y",
        "-ss", str(start),
        "-t", str(segment),
        "-i", full_wav,
        "-ac", "1",
        "-ar", "16000",
        out_wav
    ], check=True)

print("Audio clips:", clips_audio_dir)
print("Example:", sorted(os.listdir(clips_audio_dir))[:5])
```

```python
import os, glob, re
import pysrt

clips_text_dir = os.path.join(base_dir, "text_15s")
os.makedirs(clips_text_dir, exist_ok=True)

# Find Arabic SRT (could be .ar.srt or similar)
srts = sorted(glob.glob(os.path.join(video_dir, "*.srt")))
if not srts:
    raise FileNotFoundError("No .srt captions found. The video may not have Arabic captions.")
srt_path = srts[0]
print("Using captions:", srt_path)

subs = pysrt.open(srt_path, encoding="utf-8")

def to_seconds(t):
    return t.hours*3600 + t.minutes*60 + t.seconds + t.milliseconds/1000.0

def clean_caption(text):
    # remove html tags and extra spaces
    text = re.sub(r"<[^>]+>", "", text)
    text = text.replace("\n", " ").strip()
    text = re.sub(r"\s+", " ", text)
    return text

# Make a list of (start,end,text)
cap = []
for s in subs:
    st, en = to_seconds(s.start), to_seconds(s.end)
    txt = clean_caption(s.text)
    if txt:
        cap.append((st, en, txt))

segment = 15
total_segments = math.ceil(dur / segment)

for i in range(total_segments):
    start = i * segment
    end = min((i + 1) * segment, dur)
    name = f"{fmt_hhmmss(start)}_{fmt_hhmmss(end)}"
    txt_path = os.path.join(clips_text_dir, f"{name}.txt")

    # collect captions that overlap this 15s window
    lines = []
    for st, en, txt in cap:
        if en > start and st < end:   # overlap condition
            lines.append(txt)

    # de-duplicate consecutive repeats (common in captions)
    merged = []
    for line in lines:
        if not merged or merged[-1] != line:
            merged.append(line)

    with open(txt_path, "w", encoding="utf-8") as f:
        f.write("\n".join(merged).strip())

print("Text clips:", clips_text_dir)
print("Example:", sorted(os.listdir(clips_text_dir))[:5])
```

Set the desired start and end timestamps, then run the following cells.

```python
import os, subprocess

# Change these timestamps
START = '00:00:14'
END = '00:02:18'

base_dir = '.'
full_wav = os.path.join(base_dir, 'full.wav')
segment_audio = os.path.join(base_dir, 'segment.wav')

print('Segment:', START, '->', END)
```

```python
# Extract the audio segment
subprocess.run([
    'ffmpeg','-y','-i', full_wav,
    '-ss', START,
    '-to', END,
    '-ac','1','-ar','16000',
    segment_audio
], check=True)

print('Segment audio saved to:', segment_audio)
```

```python
import pysrt

srt_path = 'captions.srt'
subs = pysrt.open(srt_path, encoding='utf-8')

def to_seconds(t):
    return t.hours*3600 + t.minutes*60 + t.seconds

def parse_time(t):
    h,m,s = map(int,t.split(':'))
    return h*3600 + m*60 + s

start_sec = parse_time(START)
end_sec   = parse_time(END)

segment_text = []

for sub in subs:
    s = to_seconds(sub.start)
    e = to_seconds(sub.end)

    if not (e < start_sec or s > end_sec):
        segment_text.append(sub.text)

transcript = ' '.join(segment_text)
print('Transcript:', transcript)
```

```python
# Save transcript
text_path = 'segment.txt'
with open(text_path, 'w', encoding='utf-8') as f:
    f.write(transcript)

print('Transcript saved to:', text_path)
```

```python
import os
import re
import yt_dlp
from pydub import AudioSegment


# -----------------------------
# Download audio + subtitles
# -----------------------------
def download_video_data(url, output_dir="/kaggle/working/charity_dataset"):
    os.makedirs(output_dir, exist_ok=True)

    ydl_opts = {
        'format': 'bestaudio/best',
        'outtmpl': f'{output_dir}/%(id)s.%(ext)s',
        'writesubtitles': True,
        'writeautomaticsub': True,
        'subtitleslangs': ['ar'],
        'skip_download': False,
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(url, download=True)

    return info['id']


# -----------------------------
# Convert HH:MM:SS.xxx to ms
# -----------------------------
def time_to_ms(t):
    h, m, s = t.split(':')
    sec, ms = s.split('.')
    return (
        int(h) * 3600 * 1000
        + int(m) * 60 * 1000
        + int(sec) * 1000
        + int(ms)
    )

def timestamp_to_ms(ts):
    h, m, s = map(int, ts.split(':'))
    return (h * 3600 + m * 60 + s) * 1000
# -----------------------------
# Parse VTT subtitles
# -----------------------------
def parse_vtt(vtt_path):
    segments = []

    with open(vtt_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    i = 0
    while i < len(lines):
        line = lines[i].strip()

        if '-->' in line:
            start, end = line.split(' --> ')
            text = []

            i += 1
            while i < len(lines) and lines[i].strip():
                text.append(lines[i].strip())
                i += 1

            clean_text = ' '.join(text)
            clean_text = re.sub(r'<.*?>', '', clean_text)

            segments.append({
                "start": time_to_ms(start),
                "end": time_to_ms(end.split()[0]),
                "text": clean_text
            })

        i += 1

    return segments


# -----------------------------
# Extract audio + transcript
# -----------------------------
def extract_segment(video_id, start_time, end_time, folder="/kaggle/working/charity_dataset"):
    audio_path = None
    vtt_path = None

    # Find audio + subtitle files
    for f in os.listdir(folder):
        if f.startswith(video_id):
            if f.endswith(('.m4a', '.webm')):
                audio_path = os.path.join(folder, f)
            elif f.endswith('.ar.vtt'):
                vtt_path = os.path.join(folder, f)

    if not audio_path:
        raise FileNotFoundError("Audio file not found")

    if not vtt_path:
        raise FileNotFoundError("Arabic subtitle file not found")

    # Convert timestamps
    start_ms = timestamp_to_ms(start_time)
    end_ms = timestamp_to_ms(end_time)

    # Extract audio segment
    audio = AudioSegment.from_file(audio_path)
    segment = audio[start_ms:end_ms]

    wav_out = os.path.join(
        folder,
        f"segment_{start_time.replace(':','-')}_{end_time.replace(':','-')}.wav"
    )

    segment.export(wav_out, format="wav")

    # Extract transcript
    captions = parse_vtt(vtt_path)

    transcript = []
    for cap in captions:
        if cap['start'] < end_ms and cap['end'] > start_ms:
            transcript.append(cap['text'])

    txt_out = os.path.join(
        folder,
        f"segment_{start_time.replace(':','-')}_{end_time.replace(':','-')}.txt"
    )

    with open(txt_out, 'w', encoding='utf-8') as f:
        f.write(" ".join(transcript))

    return wav_out, txt_out
```

```python
url = "https://www.youtube.com/watch?v=_opNg-ROx70"


video_id = download_video_data(url)

START = '00:04:15'
END   = '00:08:46'

wav_file, txt_file = extract_segment(
    video_id,
    START,
    END
)

print("WAV:", wav_file)
print("TXT:", txt_file)
```
